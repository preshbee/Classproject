### Exercise 8 (8 points)

Find the list of the indexes of the array *arr = np.random.randn(10000)* for which the value of an element is greater than both the previous and the following one. And print the list of these numbers.

You should find an array that is about 1/3 long of *arr*. You can verify that the average is around zero for *arr* but it is around $0.84$ for the numbers higher than their neighbors.

```python
# here write your solution
arr = np.random.randn(100001)
list = []

for i in range(len(arr)):
    if arr[i] > arr[i-1] and arr[i] > arr[i+1]:
        list.append(arr[i])
```

```python
print(list)
print(len(list))
```

```python
print(np.average(arr))
print(np.average(list))
```
